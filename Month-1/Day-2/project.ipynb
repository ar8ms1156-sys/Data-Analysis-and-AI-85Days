### 📌 [1] البيانات الهيكلية والمصنفة من لوحة التحكم:

* **البيانات الرقمية (Numerical Data):** تشمل القيم المالية والكمية القابلة للحساب مثل: سعر المنتج (**250.0 ريال**)، كمية الطلب (**2 قطعة**)، رسوم الشحن (**28.0 ريال**)، وإجمالي الفاتورة الصافية (**528.0 ريال**).
* **البيانات الفئوية (Categorical Data):** تشمل المتغيرات النوعية المحددة مسبقاً لتصنيف الطلب مثل: طريقة الدفع (**تمارا**)، حالة الطلب (**تم الشحن**)، مدينة العميل (**الرياض**)، وفئة المنتج (**عطور**).
* **البيانات النصية الحرة (Text Data):** تشمل النصوص الحرة المرسلة من العميل وتتطلب تحليلاً نصياً متقدماً مثل: اسم العميل كامل، وعنوان التوصيل التفصيلي (حي النرجس)، الملاحظات الخاصة (طلب هدية)، وتقييم المنتج الإيجابي.

> 💡 **ملاحظة المحلل:** تم بنجاح بناء هيكلية القواميس (Dictionaries) في بايثون لفرز وتصنيف البيانات بدقة بناءً على متطلبات اليوم الثاني، مما يمهد الطريق لمعالجة البيانات وفهم سلوك المستهلك بشكل مؤتمت بالكامل.

### 🔍 [2] إجمالي الأداء والتحليل الاستراتيجي (Business Insights):

* **تحليل سلوك المتسوق السعودي:** يتبين من البيانات أن العميل اعتمد على حلول الدفع الآجل (**تمارا**)، وهو ما يعكس الاتجاه المتزايد في السوق السعودي للاعتماد على خدمات (اشترِ الآن وادفع لاحقاً - BNPL) لتعزيز القدرة الشرائية وتسهيل المعاملات.
