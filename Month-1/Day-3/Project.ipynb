```python
class ZajilModel:
    def __init__(self):
        # 1. تحديد الأسعار الثابتة لشركة زاجل
        self.base_fee = 15.0      # سعر فتح العداد (Beta 0)
        self.dist_rate = 0.5      # سعر الكيلومتر الواحد (Beta 1)
        self.weight_rate = 2.0    # سعر الكيلوجرام الواحد (Beta 2)

    def predict(self, distance, weight):
        # 2. حساب المعادلة الخطية
        total_cost = self.base_fee + (self.dist_rate * distance) + (self.weight_rate * weight)
        
        # 3. طباعة التقرير النهائي للعميل
        print("=" * 40)
        print("       📝 تقرير تكلفة الشحن الإجمالية       ")
        print("=" * 40)
        print(f"📍 المسافة المقطوعة: {distance} كم")
        print(f"⚖️ وزن الطرد الإجمالي: {weight} كجم")
        print("-" * 40)
        print(f"💰 التكلفة المتوقعة: {total_cost} ريال سعودي")
        print("=" * 40)

# --- تشغيل البرنامج وتجربته ---
app = ZajilModel()
app.predict(distance=100, weight=5)
```
